```python
import numpy as np
import pygame
import os
os.environ['SDL_VIDEODRIVER'] = 'directx'
y_size = 1327
x_y_sf = 1.78448
pygame.init()
screen = pygame.display.set_mode((int(y_size*x_y_sf), int(y_size)), pygame.DOUBLEBUF)
t_x = 0.06038
t_y = 0.10852
vel_sf = 10 ** -4
back_col = (0, 197, 125)
clock = pygame.time.Clock()
table = pygame.Rect(
    (
        t_x * y_size * x_y_sf,
        t_y * y_size,
        (1 - 2 * t_x) * x_y_sf * y_size,
        (1 - 2 * t_y + 0.002) * y_size,
    )
)
table_dim = np.array([table.width, table.height])
table_pos = np.array([table.x, table.y])
corner_pocket_size = 0.04
side_pocket_size = 0.035
pockets = {
    "upper_left": [
        (table.x, table.y + table.width * corner_pocket_size),
        (table.x, table.y),
        (table.x + table.width * corner_pocket_size, table.y),
    ],
    "upper_middle": [
        (table.x + table.width / 2 - table.width * side_pocket_size, table.y),
        (table.x + table.width / 2 + table.width * side_pocket_size, table.y),
    ],
    "upper_right": [
        (table.x + table.width, table.y + table.width * corner_pocket_size),
        (table.x + table.width, table.y),
        (table.x + table.width - table.width * corner_pocket_size, table.y),
    ],
    "lower_right": [
        (
            table.x + table.width,
            table.y + table.height - table.width * corner_pocket_size,
        ),
        (table.x + table.width, table.y + table.height),
        (
            table.x + table.width - table.width * corner_pocket_size,
            table.y + table.height,
        ),
    ],
    "lower_middle": [
        (
            table.x + table.width / 2 - table.width * side_pocket_size,
            table.y + table.height,
        ),
        (
            table.x + table.width / 2 + table.width * side_pocket_size,
            table.y + table.height,
        ),
    ],
    "lower_left": [
        (table.x, table.y + table.height - table.width * corner_pocket_size),
        (table.x, table.y + table.height),
        (table.x + table.width * corner_pocket_size, table.y + table.height),
    ],
}


class Ball(pygame.sprite.Sprite):
    ball_radius = 0.060287 / 2
    ball_size = int(y_size * 2 * ball_radius)
    yellow_ball_s = pygame.transform.smoothscale(
        pygame.image.load("yellow_ball.png"), (ball_size, ball_size)
    ).convert_alpha()
    red_ball_s = pygame.transform.smoothscale(
        pygame.image.load("red_ball.png"), (ball_size, ball_size)
    ).convert_alpha()
    rad = int(y_size * ball_radius)
    def __init__(self, pos, color, vel, *groups):
        super().__init__(*groups)
        self.image = {
            "y": self.yellow_ball_s,
            "r": self.red_ball_s,
            # "b": black_ball_s,
            # "w": white_ball_s,
        }[color]
        self.pos = pos
        self.vel = vel

    def update(self):
        dt = clock.get_time()
        dpos = self.vel * dt * vel_sf
        t_pos = rel_table(self.pos)
        screen.blit(self.image, t_pos - self.rad)
        self.pos = self.pos + dpos


def rel_table(pos):
    return np.around(table_pos + table_dim * pos).astype(int)


sprites = pygame.sprite.Group()
balls = []


def main():
    if len(balls) < 1:
        balls.append(Ball(np.array([0.2, 0.2]), "r", 5* np.array([1, 1]), sprites))
        #balls.append(Ball(np.array([0.8, 0.8]), "y", 0.7 * np.array([-1, -1]), sprites))
    for ball in balls:
        if not 0 < ball.pos[0] < 1:
            balls.pop()
            #balls.pop()
    sprites.update()
    # screen.blit(yellow_ball_s, rel_table(np.array([0.6,0.6])))
    pygame.display.flip()


background = pygame.image.load("table.png")
pygame.draw.rect(screen, (255, 255, 255), table, 1)
for pocket in pockets.values():
    pygame.draw.lines(screen, (0, 0, 255), False, pocket, 5)
background = background.convert()


fr = 100
running = True
while running:
    screen.blit(background, (0, 0))
    pygame.display.set_caption(f"{clock.get_fps()}FPS")
    clock.tick_busy_loop()
    try:
        for ev in pygame.event.get():
            if ev.type == pygame.QUIT:
                print("exiting...\n")
                running = False
        main()
    except KeyboardInterrupt:
        running = False
        pygame.quit()
    except Exception as e:
        pygame.quit()
        raise e
pygame.quit()
```
