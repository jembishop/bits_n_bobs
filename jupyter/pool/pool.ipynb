```python
import numpy as np
import pygame

y_size = 1327
x_y_sf = 1.78448
t_x = 0.06038
t_y = 0.10852
vel_sf = 10**-1
clock = pygame.time.Clock()
table = pygame.Rect(
    (
        t_x * y_size * x_y_sf,
        t_y * y_size,
        (1 - 2 * t_x) * x_y_sf * y_size,
        (1 - 2 * t_y + 0.002) * y_size,
    )
)
table_dim = np.array([table.width, table.height])
table_pos = np.array([table.x, table.y])
corner_pocket_size = 0.04
side_pocket_size = 0.035
pockets = {
    "upper_left": [
        (table.x, table.y + table.width * corner_pocket_size),
        (table.x, table.y),
        (table.x + table.width * corner_pocket_size, table.y),
    ],
    "upper_middle": [
        (table.x + table.width / 2 - table.width * side_pocket_size, table.y),
        (table.x + table.width / 2 + table.width * side_pocket_size, table.y),
    ],
    "upper_right": [
        (table.x + table.width, table.y + table.width * corner_pocket_size),
        (table.x + table.width, table.y),
        (table.x + table.width - table.width * corner_pocket_size, table.y),
    ],
    "lower_right": [
        (
            table.x + table.width,
            table.y + table.height - table.width * corner_pocket_size,
        ),
        (table.x + table.width, table.y + table.height),
        (
            table.x + table.width - table.width * corner_pocket_size,
            table.y + table.height,
        ),
    ],
    "lower_middle": [
        (
            table.x + table.width / 2 - table.width * side_pocket_size,
            table.y + table.height,
        ),
        (
            table.x + table.width / 2 + table.width * side_pocket_size,
            table.y + table.height,
        ),
    ],
    "lower_left": [
        (table.x, table.y + table.height - table.width * corner_pocket_size),
        (table.x, table.y + table.height),
        (table.x + table.width * corner_pocket_size, table.y + table.height),
    ],
}


class Ball:
    def __init__(self, pos, color, vel=np.array([0, 0])):
        self.image = {
            "y": yellow_ball_s,
            "r": red_ball_s,
            # "b": black_ball_s,
            # "w": white_ball_s,
        }[color]
        self.pos = pos
        self.vel = vel
        

    def update(self):
        dt = 1/fr
        dpos = self.vel * dt * vel_sf
        t_lpos = rel_table(self.pos)
        self.pos = self.pos + dpos
        t_pos = rel_table(self.pos)
        pygame.draw.rect(
            screen,
            (0, 197, 125),
            (t_lpos[0],  t_lpos[1], self.image.get_width(), self.image.get_height()),
            0,
        )
        screen.blit(self.image, t_pos)
        


def rel_table(pos):
    return table_pos + table_dim * pos


balls = []


def main():
    if len(balls) < 10:
        balls.append(Ball(np.random.rand(2), "r", vel=np.random.rand(2)))
    for ball in balls:
        ball.update()
    # screen.blit(yellow_ball_s, rel_table(np.array([0.6,0.6])))
    pygame.display.flip()
```

```python

background = pygame.image.load("table.png")
ball_radius = 0.060286
yellow_ball_s = pygame.transform.smoothscale(
    pygame.image.load("yellow_ball.png"), (int(y_size * ball_radius), int(y_size * ball_radius))
)
red_ball_s = pygame.transform.smoothscale(
    pygame.image.load("red_ball.png"), (int(y_size * ball_radius), int(y_size * ball_radius))
)
screen = pygame.display.set_mode((int(y_size*x_y_sf), int(y_size)))
screen.blit(background, (0, 0))
pygame.draw.rect(screen, (255, 255, 255), table, 1)
for pocket in pockets.values():
    pygame.draw.lines(screen, (0, 0, 255), False, pocket, 5)
background = background.convert()
yellow_ball_s = yellow_ball_s.convert_alpha()
red_ball_s = yellow_ball_s.convert_alpha()
pygame.init()

fr = 65
running = True
while running:
    print(clock.get_fps(), end="\r")
    clock.tick(fr)
    try:
        for ev in pygame.event.get():
            if ev.type == pygame.QUIT:
                print("exiting...\n")
                running = False
        main()
    except KeyboardInterrupt:
        running = False
        pygame.quit()
    except Exception as e:
        pygame.quit()
        raise e
pygame.quit()
```
